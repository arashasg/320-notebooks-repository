```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.scatterplot(x='year',y='proportion',data=df)
```

```python
sns.countplot(df["proportion"])
plt.xticks(rotation=45)
plt.show()
```

```python
import plotly.offline as pyo
import plotly.graph_objs as go
lowerdf = df.groupby('year').size()/df['proportion'].count()*100
labels = lowerdf.index
values = lowerdf.values

# Use `hole` to create a donut-like pie chart
fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.6)])
fig.show()
```

```python
labels1=df.proportion.value_counts().index
sizes1=df.proportion.value_counts().values
plt.figure(figsize=(11,11))
plt.pie(sizes1,labels=labels1,autopct="%1.1f%%")
plt.title("proportion",size=25)
plt.show()
```

```python
ng = df[df.proportion <250]
plt.figure(figsize=(8,4))
sns.boxplot(y="proportion",x ='year' ,data = ng)
plt.title("proportion & year < 250")
plt.xticks(rotation=45)
plt.show()
```

```python
#catplot room type and price
plt.figure(figsize=(10,6))
sns.catplot(x="year", y="proportion", data=df);
plt.xticks(rotation=90)
plt.ioff()
```

```python
datt =df[['year', 'proportion']]
g = sns.PairGrid(datt)
g = g.map(plt.scatter)
```

```python
g = sns.PairGrid(datt)
g = g.map_diag(plt.hist)
g = g.map_offdiag(plt.scatter)
```

```python
g = sns.PairGrid(datt, hue="year")
g = g.map_diag(plt.hist)
g = g.map_offdiag(plt.scatter)
g = g.add_legend()
```

```python
g = sns.PairGrid(datt)
g = g.map_upper(plt.scatter)
g = g.map_lower(sns.kdeplot, cmap="Blues_d")
g = g.map_diag(sns.kdeplot, lw=3, legend=False)
```

```python
g = sns.JointGrid(x="year", y="proportion", data=df)
g = g.plot(sns.regplot, sns.distplot)
```

```python
g = sns.JointGrid(x="year", y="proportion", data=df, space=0)
g = g.plot_joint(sns.kdeplot, cmap="Blues_d")
g = g.plot_marginals(sns.kdeplot, shade=True)
```

```python
g = sns.JointGrid(x="year", y="proportion", data=df, height=5, ratio=2)
g = g.plot_joint(sns.kdeplot, cmap="Reds_d")
g = g.plot_marginals(sns.kdeplot, color="r", shade=True)
```
