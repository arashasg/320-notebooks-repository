Next we have done linear regression to predict the GLD variable value based on other variables.
We have mentioned the predictor and explanatory variables and split the entire data set into train dataset to train the machine learning model and to apply it on test data set.”train_test_split ” function does the job of splitting the train and test dataset.LinearRegression() function is applied on the train values of the variables. In the fitted model the coefficients come as SPX(0.006584),                                   USO(-0.255845),SLV(3.1477),EUR/USD(-8.9691). Clearly EUR/USD variable has a negative impact on the increase of the values of the variable GLD. SLV has a fairly positive effect.The impact of SPX and USO is merely unmentionable.
Next we have predicted the values of Test data based on the trained algorithm with the function lm.predict and obtained the values of the variable GLD as “predictions”.Next we have compared both the predicted values iven values of test in a scatterplot.
The scatterplot clearly shows that the deviation between the given values or observed values of GLD and predicted values of GLD (here named as “predictions”) being maximum at the large values of the variable GLD.


```python
X = gold_data[['SPX','USO','SLV','EUR/USD']]  #mentioning the predictor variables 
y = gold_data['GLD']   # mentioning response variable
#splitting the train and test dataset
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)  #splitting train and test dataset
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)  #training the model with the linear regression function
```

```python
lm.coef_  #getting the coefficients after the regression
```

```python
print(lm.intercept_) #printing the intercept
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df  #coefficients in a matrix form
```

```python
predictions = lm.predict(X_test)
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y') #
```

```python
plt.plot(y_test,predictions,linestyle='solid')
```

The errors for this regression fitting comes along as 
MAE :   5.938773207131296  (Mean absolute error)
MSE :   63.54693696387926  (Mean Square Error)  (explains the deviation)
RMAE :   7.971633268275659 (Root Mean Square Error)



```python
from sklearn import metrics
print('MAE :'," ", metrics.mean_absolute_error(y_test,predictions))
print('MSE :'," ", metrics.mean_squared_error(y_test,predictions))
print('RMAE :'," ", np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```
