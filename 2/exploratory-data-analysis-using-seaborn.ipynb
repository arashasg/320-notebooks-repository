- country-year…Combination of country and year.    
- HDI for year…The Human Development Index (HDI) is a simple indicator for measuring the average degree of accomplishment in a country regarding three aspects of human development: health, education, and income. 
- generation…Generations in America.  
    - G.I. Generation…1900 - 1924  
    - Silent…1925 - 1942  
    - Boomers…1946 - 1964  
    - Generation X …1965 - 1980    
    - Millenials…1981 - 2000  
    - Generation Z…2001 - 2010

# Visualization

First of all, I'll look at the number of data in each country.  
**It is the number of data, not the number of suicides.  **

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,50))
y = data['country']
sns.set_context("paper", 2.5, {"lines.linewidth": 4})
sns.countplot(y=y,label='count')
```

Next, let's look at the number of data each year.  

```python
plt.figure(figsize=(30,10))
y = data['year']
sns.set_context("paper", 2.0, {"lines.linewidth": 4})
sns.countplot(y,label='count')
```

Let's check the number of data for each gender.  

```python
y = data['sex']
sns.set_context("paper", 2.5, {"lines.linewidth":4})
sns.countplot(y,label='count')
```

The number of data for each gender is almost the same.  
Next is the number of data each age.  

```python
plt.figure(figsize=(15,5))
y = data['age']
sns.set_context("paper", 2.0, {"lines.linewidth": 4})
sns.countplot(y,label='count',order=['5-14 years','15-24 years','25-34 years','35-54 years','55-74 years','75+ years'])
```

The number of data for each age is almost the same too.  
Perhaps it is considered that all gender and age combinations are included in country-year.  

```python
plt.figure(figsize=(15,5))
y = data['generation']
sns.set_context("paper", 2.0, {"lines.linewidth": 4})
sns.countplot(y,label='generation',order=['G.I. Generation','Silent','Boomers','Generation X','Millenials','Generation Z'])
```

## Correlation between suicides/100k_pop and HDI

```python
data_scaled = data.loc[:,['HDI_for_year','suicides/100k_pop']]
data_scaled = (data_scaled - data_scaled.mean()) / data_scaled.std()
plt.figure(figsize=(10,10))
sns.scatterplot(data=data_scaled,x='HDI_for_year',y='suicides/100k_pop')
```

It seems that there is not much correlation.

## Correlation between suicides/100k_pop and GDP for year

```python
data_scaled = data.loc[:,['gdp_for_year($)','suicides/100k_pop']]
data_scaled = (data_scaled - data_scaled.mean()) / data_scaled.std()
sns.scatterplot(data=data_scaled,x='gdp_for_year($)',y='suicides/100k_pop')
```

## Correlation between suicides/100k_pop and GDP per capita

```python
data_scaled = data.loc[:,['gdp_per_capita($)','suicides/100k_pop']]
data_scaled = (data_scaled - data_scaled.mean()) / data_scaled.std()
sns.scatterplot(data=data_scaled,x='gdp_per_capita($)',y='suicides/100k_pop')
```

When GDP is low, it seems that there will be more suicides per 100,000 people.
