* We use correlation map to understand the relation between variables inside our data.
* If we have 2 features and the correrlation is 1, it means these variables have positive correlation between each other**. 


```python
print(pokemon_data.corr())
```

* With "corr()" method, we can see the correlation between variables.
* These correlation can be between -1 and 1.
* Also we can show these correlations using heat map.( Colorfull version) 
* Correlation between 0 and 1 => positive coorelation
* Correlation between -1 and 0 => negative correlation

```python
f, ax = plt.subplots(figsize = (10,10)) # allcocating figure size
sns.heatmap(pokemon_data.corr(), annot = True, linewidths = .5, fmt = ".1f", ax=ax)
plt.show()
```

* annot True -> for making numbers visible.
* linewidth -> width between squares.
* fnt -> After dot, how many numbers can we see

# LINE PLOT

```python
plt.figure(2)
pokemon_data.Defense.plot(kind = "line", color = "orange", grid = True, linestyle = ":")
pokemon_data.Speed.plot(color = "green", grid = True, linestyle = ":")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Speed-Defense Line Plot")
plt.legend()
plt.show()
```

# SCATTER PLOT

```python
pokemon_data.plot(kind = "scatter", x = "Attack", y = "Defense", color = "orange")
plt.xlabel("Attack")
plt.ylabel("Defence")
plt.title("Attack vs Defense Scatter Plot")
plt.legend()
plt.show()
```

# HISTOGRAM

```python
plt.figure(4)
pokemon_data.Speed.plot(kind = "hist", color = "green", bins = 80)
plt.xlabel("Speed")
plt.ylabel("Frequency")
plt.title("Pokemon Data Speed Visualization")
plt.show()
```

**Explotary Data Analysis (EDA)**

* Median (quantile 50%) -> middle number of the list
* Lower quartile (quantile 25%)-> the median between the smallest number and the median.
* Upper quartile (quantile 75%) -> the median between the largest number and the median.

* Outliers (Aykırı): If a value is very higher or lower from remaining part of the list.

* IQR(Outlier) = | Upper quantile - Lower quantile|

* "describe()" method gives statistical values of the data
